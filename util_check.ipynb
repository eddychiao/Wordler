```python
import utils

word_file = 'wordle-words.txt'
num_words = len(utils.listify(word_file))

# all words
word_list = utils.listify(word_file)
# all words with unique letters
unique_word_list = utils.unique_listify(word_file)

print(len(word_list))
print(len(unique_word_list))

word_list
```

```python
freq, pos = utils.letter_dists(word_file)

freq_percent = utils.freq_dists_percent(freq, num_words)
pos_percent = utils.pos_dists_percent(pos, num_words)

freq_percent
```

```python
pos_percent
```

```python
freq_scores = utils.freq_score_all_words(word_list, freq_percent)
freq_scores_unique = utils.freq_score_all_words(unique_word_list, freq_percent)

pos_scores = utils.pos_score_all_words(word_list, pos_percent)
pos_scores_unique = utils.pos_score_all_words(unique_word_list, pos_percent)

# print(freq_scores_unique)

print(utils.lookup_score('slats', freq_scores))
print(utils.lookup_score('slats', pos_scores))

print()

print(utils.lookup_score('orate', freq_scores_unique))
print(utils.lookup_score('stare', freq_scores_unique))

print()

print(utils.lookup_score('orate', pos_scores_unique))
```
